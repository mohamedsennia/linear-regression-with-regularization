# Transformation de données
Dans cette partie, nous aimerions transformer nos données afin d'avoir une fonction polynomiale de degrée 6.

La fonction sera:

$x_0=1$

$x_1=x_1$

$x_2=x_2$

$x_3={x_1}^2$

$x_4={x_1}x_2$

$x_5={x_2}^2$

$x_6={x_1}^3$

$x_7={x_1}^2 x_2$

$x_8= x_1 {x_2}^2$

$x_9= {x_2}^3$

...


Pour un polynme de degrée 6 à 2 variables nous aurons 28 caracteristiques

**Question**: comment avons nous trouvé ce chiffre?

**Astuce**: référez vous aux probabilités

```python
def mapping(X):
    
    cols = 28
    degree=7
    outX= np.ones((X.shape[0],cols))
    X1=X[:,1]
    X2=X[:,2]
    k=0
    for i in range(degree):
        for j in range(i+1):
            outX[:, k] = np.power(X1,i-j)*(np.power(X2,j));
            k=k+1
    return outX    
```

```python
X2=mapping(X)
X2.shape
```

# Descente du Gradient : Préparation des fonctions

0- Fonction mpgistique (Sigmoid)

```python
def Sigmoid(z):
    # pour une valeur donnée, cette fonction calculera sa sigmoid
    return 1/(1+np.exp(-z));
 
```

```python
k=Sigmoid(-10)
k
```

1- Calcul du coût

Cette fonction servira à calculer le cout $J(\theta_0,\theta_1)$

Elle prendra l'ensemble de données d'apprentissage en entrée ainsi que les paramètres définis initialement

```python
def computeCost(X, y, theta):
    z=X.dot(theta)
    j=0
    m=X.shape[0]
    for i in range(m):
        h=Sigmoid(z[i])
        j=j+(y[i]*math.log(h))+((1-y[i])*math.log(1-h))
    j=-j/m
    return j
n=X2.shape[1]
theta = np.zeros((n, 1))
def computeCostReg(X, y, theta,lambdaa):
    z=X.dot(theta)
    j=0
    m=X.shape[0]
    for i in range(m):
        h=Sigmoid(z[i])
        j=j+(y[i]*math.log(h))+((1-y[i])*math.log(1-h))
    n=theta.shape[0]
    s=0
    for k in range(1,n):
        s=s+math.pow(theta[k,0],2)
    s=lambdaa*s
    j=j+s
    j=-j/(2*m)
    return j
```

2- Fonction de la descente du gradient

Cette fonction mettra à jour les paramètres $\theta_0,\theta_1$ jusqu'à convergence: atteinte du nombre d'itérations max, ou dérivée assez petite.

```python
def gradientDescent(X, y, theta, alpha, iterations):
    # garder aussi le cout à chaque itération 
    # pour afficher le coût en fonction de theta0 et theta1
    n=X.shape[1]
    m=X.shape[0]
    h=np.zeros(n)
    s=np.zeros(n)
    z=X.dot(theta)
    for i in range(iterations):
        h=np.zeros(n)
        s=np.zeros(n)
        z=X.dot(theta)
        for j in range(m):
            for k in range(n):
                h[k]=(Sigmoid(z[j])-y[j])*X[j][k]
                s[k]=s[k]+h[k]
        for k in range(n):
            s[k]=s[k]/m
            theta[k]=theta[k]-alpha*s[k]
    return theta
def gradientDescentWReg(X, y, theta, alpha,lambdaa, iterations):
    # garder aussi le cout à chaque itération 
    # pour afficher le coût en fonction de theta0 et theta1
    costs = np.zeros(iterations)
    n=X.shape[1]
    m=X.shape[0]
    h=np.zeros(n)
    s=np.zeros(n)
    z=X.dot(theta)
    for i in range(iterations):
        h=np.zeros(n)
        s=np.zeros(n)
        z=X.dot(theta)
        for j in range(m):
            for k in range(n):
                h[k]=(Sigmoid(z[j])-y[j])*X[j][k]
                s[k]=s[k]+h[k]
        for k in range(n):
            s[k]=s[k]/m
            if k>0:
                s[k]=s[k]+(lambdaa/m)*theta[k]
            theta[k]=theta[k]-alpha*s[k]
        costs[i] = computeCost(X,y, theta)
    return theta,costs
```

# Dessin de la limite de decision (Descision Boundary)
Dans cette partie, nous aimerions dessiner la ligne separatrice d nos données

```python
def drawCircle(X,theta_final):
    x1_min, x1_max = X[:, 1].min(), X[:, 1].max(),
    x2_min, x2_max = X[:, 2].min(), X[:, 2].max(),
    xx1, xx2 = np.meshgrid(np.linspace(x1_min, x1_max), np.linspace(x2_min, x2_max))

    h = Sigmoid(map_feature(xx1.reshape(-1), xx2.reshape(-1)).dot(theta_final))
    h = h.reshape(xx1.shape)
    plt.contour(xx1, xx2, h, [0.5], linewidths=1, colors='g');
    plt.show()


def map_feature(x1, x2):
    x1.shape =(x1.size,1)
    x2.shape =(x2.size,1)
    degree =6
    mapped_fea = np.ones(shape=(x1[:,0].size,1))
    for i in range(1, degree +1):
        for j in range(i +1):
            r =(x1 **(i - j))*(x2 ** j)
            mapped_fea = np.append(mapped_fea, r, axis=1)
    return mapped_fea
```

# Classification (Prédiction)
Ici il serait interessant de calculer la prédiction en utilisant un seuil
i.e. si h>seuil alors classe =1 sinon classe = 0

```python
def predict(X):
    y_pred = np.zeros((X.shape[0], 1))
    h = Sigmoid(X.dot(theta))

    for i in range(X.shape[0]):
        if  h[i] >0.5:
            y_pred[i] =1
        else:
            y_pred[i] =0
    return y_pred
```
